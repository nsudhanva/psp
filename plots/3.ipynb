```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('3.xlsx')
```

```python
df.head()
```

```python
id_rms_df = df.groupby(['ID', 'ID.1'])['RMS Value'].apply(list).to_frame()
id_rms_df = id_rms_df.reset_index()
```

```python
bins = np.arange(0, df['RMS Value'].max())

for index, row in id_rms_df.iterrows():
#     print(row['ID'], row['ID.1'], row['RMS Value'])
    plt.hist(row['RMS Value'], bins=bins, edgecolor='black', linewidth=1.2)
    plt.xlabel(row['ID'] + ', ' + row['ID.1'] + ' -- RMS -- ')
    plt.title('RMS Distribution of ' + row['ID'] + ', ' + row['ID.1'])
    plt.show()
```
